Naive Bayes Classifier

```python
train_data = df[df["TrainData"] == 0]
test_data = df[df["TrainData"] == 1]

```

```python
#Setting Parameters & Training Gaussian Distribution
disease_train_data = train_data[train_data["Dementia"] == 1]
healthy_train_data = train_data[train_data["Dementia"] == 0]

right_mean_disease = statistics.mean(disease_train_data["RightHippoVol"])
right_mean_healthy = statistics.mean(healthy_train_data["RightHippoVol"])
right_var_disease = statistics.variance(disease_train_data["RightHippoVol"])
right_var_healthy = statistics.variance(healthy_train_data["RightHippoVol"])

left_mean_disease = statistics.mean(disease_train_data["LeftHippoVol"])
left_mean_healthy = statistics.mean(healthy_train_data["LeftHippoVol"])
left_var_disease = statistics.variance(disease_train_data["LeftHippoVol"])
left_var_healthy = statistics.variance(healthy_train_data["LeftHippoVol"])
```

```python

# P(X* | C) likelihood function 
def likelihood(left_hip_testpoint, right_hip_testpoint):
    
    pdf_left_disease = norm.pdf(left_hip_testpoint, loc = left_mean_disease, scale = np.sqrt(left_var_disease))
    pdf_right_diseases = norm.pdf(right_hip_testpoint, loc = right_mean_disease, scale = np.sqrt(right_var_disease))  
    pdf_disease = pdf_left_disease*pdf_right_diseases


    
    pdf_left_healthy = norm.pdf(left_hip_testpoint, loc = left_mean_healthy, scale = np.sqrt(left_var_healthy))
    pdf_right_healthy = norm.pdf(right_hip_testpoint, loc = right_mean_healthy, scale = np.sqrt(right_var_healthy))
    pdf_healthy = pdf_left_healthy*pdf_right_healthy


    pdf = [ pdf_healthy, pdf_disease,]
    return pdf

print( likelihood(3052, 2770))

    
```

```python
# Given Ck P(Ck)
def prior(clas):
    clas_amt = sum(train_data["Dementia"] == clas)
    return clas_amt / len(train_data["Dementia"])

prior(0)
        
```

```python
# P(x*)
def total_probability(left_hip_testpoint, right_hip_testpoint):
    likely = likelihood(left_hip_testpoint,right_hip_testpoint)
    total = likely[0]* prior(0) + likely[1] *prior(1)
    return total

print(total_probability(3052, 2770))
    
```

```python
# P(C | x) 
def prediction(left_hip_testpoint, right_hip_testpoint):


    pred_value_disease = (likelihood(left_hip_testpoint,right_hip_testpoint)[1] * prior(1))/\
    total_probability(left_hip_testpoint, right_hip_testpoint)

    if(pred_value_disease>.5):
        return 1
    else:
        return 0 
        
```

```python
def two(current_val):
    if current_val == 2:
        new_value = current_val - 1
    elif current_val == 0:
        new_value = current_val + 1
    else:
        new_value = current_val - 1
    return new_value
          

```

```python
# Running Through the Testing Data 
test_data.loc[:, "ClassifierResults"] = test_data.apply(lambda row: prediction(row["RightHippoVol"], row["LeftHippoVol"]), axis=1)
test_data["Accuracy"] = test_data["Dementia"] + test_data["ClassifierResults"]
test_data['Accuracy'].apply(two)

#calculating total number of similarities
total_sum = 0
for value in test_data["Accuracy"]:
    if value == 2 or value == 0:
        total_sum += 1
        
# Calculate Accuracy
accuracy = total_sum / len(test_data)
print(accuracy)
#model was around 77.68% accurate
```
